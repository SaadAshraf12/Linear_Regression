```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
companies = pd.read_csv('1000_Companies.csv')
```

```python
# Separate the 'State' column from the dataset
X = companies.iloc[:, :-2].values  # Exclude the last two columns (State and Profit)
Y = companies.iloc[:, -1].values  # The last column (Profit)
```

```python
# Mean X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Total number of values
m = len(X)

# Using the formula to calculate b1 and b0
numer = 0
denom = 0

for i in range(m):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2

b1 = numer / denom  # Slope (m)
b0 = mean_y - (b1 * mean_x)  # Intercept (c)

# Print coefficients
print(b1, b0)
```

```python
# Ploting Line
max_x = np.max(X) + 100
min_x = np.min(X) - 100

x = np.linspace(min_x, max_x, 100)
y = b0 + b1 * x

plt.rcParams['figure.figsize'] = (10.0, 5.0)

# Ploting Line
plt.plot(x, y, color='blue', label='Regression Line')

# Ploting Scatter Points
plt.scatter(X, Y, c='green', label='Scatter data')

plt.xlabel('R&D Spend')
plt.ylabel('Profit')
plt.legend()
plt.show()
```

```python
y_pred = b0 + b1 * X
ss_tot = np.sum((Y - mean_y) ** 2)
ss_res = np.sum((Y - y_pred) ** 2)
r2 = 1 - (ss_res / ss_tot)
print("R2 Score:", r2)
```
